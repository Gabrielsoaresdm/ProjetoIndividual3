# Import relatÃ³rio

```python
import pandas as pd
# from matplotlib import pyplot as plt

base_relatorio = pd.read_excel("relatorio.xlsx")
base_relatorio
```


# Total e MÃ©dia de Horas Trabalhadas

```python
# TOTAL DE HORAS TRABALHADAS
total_horas = base_relatorio['Horas Trabalhadas'].sum()
print("O total de horas trabalhadas da equipe Ã© de: " + str(total_horas) + " horas")
```

```python
# MÃ‰DIA DE HORAS TRABALHADAS
media_horas = int(base_relatorio['Horas Trabalhadas'].mean().round())
print("A mÃ©dia de horas trabalhadas da equipe Ã© de: " + str(media_horas) + " horas")
```

```python
#import da biblioteca matplotlib
from matplotlib import pyplot as plt

# Criando o grÃ¡fico de barras
ax = base_relatorio.plot(kind='bar', title='Horas trabalhadas', x='Dia', y='Horas Trabalhadas')

# Definindo os limites do eixo y para exibir apenas valores inteiros
ax.set_ylim(bottom=0, top=base_relatorio['Horas Trabalhadas'].max() + 1)

# Ocultando as bordas superior e direita do grÃ¡fico corretamente
plt.gca().spines[['top', 'right',]].set_visible(False)

# Removendo a legenda
ax.legend().set_visible(False)

# Adicionando rÃ³tulos de dados no topo das colunas
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')


# Adicionando o texto informativo diretamente ao objeto de eixo
plt.axhline(y=media_horas, color='r', linestyle='--', label=f'MÃ©dia de {media_horas} horas trabalhadas por dia.')


# Exibindo o grÃ¡fico
plt.legend(bbox_to_anchor=(0.7, 1), loc='upper left')
plt.show()

```

# Total e MÃ©dia de Tarefas ConcluÃ­das

---



```python
# TOTAL DE TAREFAS CONCLUIDAS
total_tarefas = base_relatorio['Tarefas ConcluÃ­das'].sum()
print("O total de tarefas concluÃ­das da equipe Ã© de: " + str(total_tarefas) + " tarefas")
```

```python
# MÃ‰DIA DE TAREFAS CONCLUIDAS
media_tarefas = int(base_relatorio['Tarefas ConcluÃ­das'].mean())
print("A mÃ©dia de tarefas concluÃ­das da equipe Ã© de: " + str(media_tarefas) + " tarefas")
```

```python
#import da biblioteca matplotlib
from matplotlib import pyplot as plt

# Criando o grÃ¡fico de barras
ax = base_relatorio.plot(kind='bar', title='Tarefas ConcluÃ­das', x='Dia', y='Tarefas ConcluÃ­das')

# Definindo os limites do eixo y para exibir apenas valores inteiros
ax.set_ylim(bottom=0, top=base_relatorio['Tarefas ConcluÃ­das'].max() + 1)

# Ocultando as bordas superior e direita do grÃ¡fico corretamente
plt.gca().spines[['top', 'right',]].set_visible(False)

# Removendo a legenda
ax.legend().set_visible(False)

# Adicionando rÃ³tulos de dados no topo das colunas
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# Adicionando o texto informativo diretamente ao objeto de eixo
plt.axhline(y=media_tarefas, color='r', linestyle='--', label=f'MÃ©dia de {media_tarefas} bugs concluÃ­das por dia.')


# Exibindo o grÃ¡fico
plt.legend(bbox_to_anchor=(0.7, 1), loc='upper left')
plt.show()
```

# Tarefas ConcluÃ­das por hora (Produtividade DiÃ¡ria)

```python
# TAREFAS CONCLUÃDAS POR HORA (PRODUTIVIDADE DIÃRIA)
tarefas_hora = total_tarefas/total_horas
print("A produtividade diÃ¡ria da equipe Ã© de: " + str(tarefas_hora.round(1)) + " tarefas por hora")

# Calcular a mÃ©dia
media = (base_relatorio['Tarefas ConcluÃ­das'] / base_relatorio['Horas Trabalhadas']).round(1)
print(media)
```

```python
# Importar a biblioteca matplotlib
import matplotlib.pyplot as plt

# Criar o grÃ¡fico de linha
plt.plot(base_relatorio['Dia'], media)

# Definir tÃ­tulo e rÃ³tulos dos eixos
plt.title('Taxa de Produtividade DiÃ¡ria')
plt.xlabel('Dia')
plt.ylabel('Taxa de Produtividade')

# Ocultar as bordas superior e direita do grÃ¡fico corretamente
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Arredondando o nÃºmero
tarefas_hora_arredondado = round(tarefas_hora, 1)

# Adicionando a linha horizontal com o texto informativo
plt.axhline(y=tarefas_hora, color='r', linestyle='--', label=f'MÃ©dia de {tarefas_hora_arredondado} bugs concluÃ­das por dia.')

# Exibindo o grÃ¡fico
plt.legend(bbox_to_anchor=(0.7, 1), loc='upper left')
plt.show()
```
