# Import relatório

```python
import pandas as pd
# from matplotlib import pyplot as plt

base_relatorio = pd.read_excel("relatorio.xlsx")
base_relatorio
```


# Total e Média de Horas Trabalhadas

```python
# TOTAL DE HORAS TRABALHADAS
total_horas = base_relatorio['Horas Trabalhadas'].sum()
print("O total de horas trabalhadas da equipe é de: " + str(total_horas) + " horas")
```

```python
# MÉDIA DE HORAS TRABALHADAS
media_horas = int(base_relatorio['Horas Trabalhadas'].mean().round())
print("A média de horas trabalhadas da equipe é de: " + str(media_horas) + " horas")
```

```python
#import da biblioteca matplotlib
from matplotlib import pyplot as plt

# Criando o gráfico de barras
ax = base_relatorio.plot(kind='bar', title='Horas trabalhadas', x='Dia', y='Horas Trabalhadas')

# Definindo os limites do eixo y para exibir apenas valores inteiros
ax.set_ylim(bottom=0, top=base_relatorio['Horas Trabalhadas'].max() + 1)

# Ocultando as bordas superior e direita do gráfico corretamente
plt.gca().spines[['top', 'right',]].set_visible(False)

# Removendo a legenda
ax.legend().set_visible(False)

# Adicionando rótulos de dados no topo das colunas
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')


# Adicionando o texto informativo diretamente ao objeto de eixo
plt.axhline(y=media_horas, color='r', linestyle='--', label=f'Média de {media_horas} horas trabalhadas por dia.')


# Exibindo o gráfico
plt.legend(bbox_to_anchor=(0.7, 1), loc='upper left')
plt.show()

```

# Total e Média de Bugs Corrigidos

```python
# TOTAL DE BUGS CORRIGIDOS
total_bugs = base_relatorio['Bugs Corridos'].sum()
print("O total de Bugs corrigidos pela equipe é de: " + str(total_bugs) + " Bugs")
```

```python
# MÉDIA DE BUGS CORRIGIDOS
media_bugs = int(base_relatorio['Bugs Corridos'].mean())
print("A média de Bugs corrigidos pela equipe é de: " + str(media_bugs) + " Bugs")
```

```python
#import da biblioteca matplotlib
from matplotlib import pyplot as plt

# Criando o gráfico de barras
ax = base_relatorio.plot(kind='bar', title='Bugs Corridos', x='Dia', y='Bugs Corridos')

# Definindo os limites do eixo y para exibir apenas valores inteiros
ax.set_ylim(bottom=0, top=base_relatorio['Bugs Corridos'].max() + 1)

# Ocultando as bordas superior e direita do gráfico corretamente
plt.gca().spines[['top', 'right',]].set_visible(False)

# Removendo a legenda
ax.legend().set_visible(False)

# Adicionando rótulos de dados no topo das colunas
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# Adicionando o texto informativo diretamente ao objeto de eixo
plt.axhline(y=media_bugs, color='r', linestyle='--', label=f'Média de {media_bugs} bugs corrigidos por dia.')


# Exibindo o gráfico
plt.legend(bbox_to_anchor=(0.7, 1), loc='upper left')
plt.show()
```

# Total e Média de Tarefas Concluídas

---



```python
# TOTAL DE TAREFAS CONCLUIDAS
total_tarefas = base_relatorio['Tarefas Concluídas'].sum()
print("O total de tarefas concluídas da equipe é de: " + str(total_tarefas) + " tarefas")
```

```python
# MÉDIA DE TAREFAS CONCLUIDAS
media_tarefas = int(base_relatorio['Tarefas Concluídas'].mean())
print("A média de tarefas concluídas da equipe é de: " + str(media_tarefas) + " tarefas")
```

```python
#import da biblioteca matplotlib
from matplotlib import pyplot as plt

# Criando o gráfico de barras
ax = base_relatorio.plot(kind='bar', title='Tarefas Concluídas', x='Dia', y='Tarefas Concluídas')

# Definindo os limites do eixo y para exibir apenas valores inteiros
ax.set_ylim(bottom=0, top=base_relatorio['Tarefas Concluídas'].max() + 1)

# Ocultando as bordas superior e direita do gráfico corretamente
plt.gca().spines[['top', 'right',]].set_visible(False)

# Removendo a legenda
ax.legend().set_visible(False)

# Adicionando rótulos de dados no topo das colunas
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

# Adicionando o texto informativo diretamente ao objeto de eixo
plt.axhline(y=media_tarefas, color='r', linestyle='--', label=f'Média de {media_tarefas} bugs concluídas por dia.')


# Exibindo o gráfico
plt.legend(bbox_to_anchor=(0.7, 1), loc='upper left')
plt.show()
```

# Tarefas Concluídas por hora (Produtividade Diária)

```python
# TAREFAS CONCLUÍDAS POR HORA (PRODUTIVIDADE DIÁRIA)
tarefas_hora = total_tarefas/total_horas
print("A produtividade diária da equipe é de: " + str(tarefas_hora.round(1)) + " tarefas por hora")

# Calcular a média
media = (base_relatorio['Tarefas Concluídas'] / base_relatorio['Horas Trabalhadas']).round(1)
print(media)
```

```python
# Importar a biblioteca matplotlib
import matplotlib.pyplot as plt

# Criar o gráfico de linha
plt.plot(base_relatorio['Dia'], media)

# Definir título e rótulos dos eixos
plt.title('Taxa de Produtividade Diária')
plt.xlabel('Dia')
plt.ylabel('Taxa de Produtividade')

# Ocultar as bordas superior e direita do gráfico corretamente
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Arredondando o número
tarefas_hora_arredondado = round(tarefas_hora, 1)

# Adicionando a linha horizontal com o texto informativo
plt.axhline(y=tarefas_hora, color='r', linestyle='--', label=f'Média de {tarefas_hora_arredondado} bugs concluídas por dia.')

# Exibindo o gráfico
plt.legend(bbox_to_anchor=(0.7, 1), loc='upper left')
plt.show()
```
